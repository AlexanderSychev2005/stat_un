## Mean, Mediad, and Mode

```python
print("Mean of sleep hours:", statistics.mean(dataset['Sleep Hours']))
print("Median:", statistics.median(dataset['Sleep Hours']))
print("Mode:", statistics.mode(dataset['Sleep Hours']))
```

## Variance and standard deviation

```python
variance = statistics.pvariance(dataset['Sleep Hours'])
std_deviation = statistics.pstdev(dataset['Sleep Hours'])
print("Variance", variance)
print("Standard deviation", std_deviation)
```

## Format data

```python
info = pd.DataFrame([["Alex", "380987689876", "alex@ukr.net"],
                    ["Ighor", "380678456522", "ighor@..."]], columns=["Name", "Phone", "Email"])
info
```

```python
import re
def get_formatted_phone(value):
  result = re.fullmatch(r'(380)(\d{2})(\d{3})(\d{2})(\d{2})', value)
  return "-".join(result.groups()) if result else False

formatted_phone = info['Phone'].map(get_formatted_phone)
```

```python
info['Phone'] = formatted_phone
info
```
