## 7. Контрольна карта середніх арифметичних


```python

# Розбиваємо дані на групи (наприклад, по 5 значень у кожній групі)
group_size = 5
num_groups = len(dataset1) // group_size
group_means = [np.mean(dataset1[i * group_size:(i + 1) * group_size]) for i in range(num_groups)]

# Границі контролю
UCL = M  + 3 * np.sqrt(D / group_size)  # Верхня контрольна межа
LCL = M  - 3 * np.sqrt(D / group_size)  # Нижня контрольна межа

# Графік контрольної карти
plt.figure(figsize=(10, 5))
plt.plot(range(1, num_groups + 1), group_means, marker='o', linestyle='-', color='b', label='Середні значення груп')
plt.axhline(y=M , color='g', linestyle='--', label='Середнє значення')
plt.axhline(y=UCL, color='r', linestyle='--', label='Верхня контрольна межа (UCL)')
plt.axhline(y=LCL, color='r', linestyle='--', label='Нижня контрольна межа (LCL)')
plt.xlabel('Групи')
plt.ylabel('Середнє значення')
plt.title('Контрольна карта середніх арифметичних')
plt.legend()
plt.grid(True)
plt.show()

# Висновок: Чи виходять середні значення за контрольні межі?
out_of_control = any(x < LCL or x > UCL for x in group_means)
print(out_of_control)
```
